```python
import csv,json
import pandas as pd
#DEFININCIONES DE VARIABLES
cabecera = []
cabecera_analisis = []
registros = []
x=[]
y=[]
z=[]
datos = []

#OBTENIENDO DATOS DEL CSV ORIGINAL
with open('GLOBANT.csv', 'r') as csvfile:
    lector = csv.reader(csvfile, delimiter = ",")
    cabecera = next(lector)
    for fila in lector:
      registros.append(fila)
# creacion de las listas
cabecera_analisis=["fecha","comportamiento de la accion","punto medio HIGH-LOW"] 
for i in range(len(registros)): 
  x.append(registros[i][0])
  if  float(registros[i][4]) - float(registros[i][1]) > 0:
    j="SUBE" 
  if float(registros[i][4]) - float(registros[i][1]) < 0:
    j="BAJA"
  if float(registros[i][4]) - float(registros[i][1])== 0:
    j="ESTABLE"
  y.append(j)
  z.append(((float(registros[i][2]) - float(registros[i][3]))/2))
 
datos = [list(z) for z in list(zip(x,y,z))]
print(datos)
```

```python
with open("analisis_archivo.csv","w") as salida:
  writer=csv.writer(salida,delimiter="\t")
  writer.writerow(cabecera_analisis)
  writer.writerows(datos)
```
