```python
#1. Dado un nÃºmero entero, determinar si es par o impar

""" para resolver este problema se utiliza la operacion modulo (%) que nos devuelve el 
residuo de una division, en este caso se divide entre 2.
Ejemplo:  6/2 = 3 y queda 0 de resido; por lo tanto 6 es par.
         19/2 = 9 y queda de residuo 1; por lo tanto 19 es impar."""

num = int(input("escribe un numero: "))#input ingresar datos, si es numerico o literal el interprete lo ve como un texto.
                                       #int convierte el texto a un dato numerico entero.
if num % 2 == 0:                       #num % 2 nos da el residuo de la operacion num/2, ahora (num % 2 == 0) es una expresion que valida si residuo es cero o no.
  print ("el numero es par")           # if --> valida la condicion ( num % 2 == 0 ); si la condicion es verdadera se ejecuta la linea siguete ( impriir el numero es par), 
else: print ("el numero es impar")     # si es falsa no se ejecuta la linea anterior; sino la que esta despues de ella.


```

```python
# 2.  clasifica un triangulo segun el numero de lados.

""" Solicitar la longitud de los tres lados de un triÃ¡ngulo, clasificarlos segÃºn el tamaÃ±o
 de sus Compruebe que se puede dibujar un triÃ¡ngulo, aplicando el teorema de la desigualdad 
 del triÃ¡ngulo, el cual establece que la suma de dos lados del triÃ¡ngulo siempre es mayor que la
 medida del tercer lado, en caso de no cumplirlo sacar un mensaje.  """

x = int(1)
lado_a = int(input ("ingrese el primer lado "))
lado_b = int(input ("ingrese el segundo lado "))
lado_c = int(input ("ingrese el tercer lado "))

if lado_a > lado_b:
  x=lado_b; 
  lado_b = lado_a; 
  lado_a = x;

if lado_a > lado_c:
  x=lado_c; 
  lado_c = lado_a; 
  lado_a = x;

if lado_b > lado_c:
  x=lado_c; 
  lado_c = lado_b; 
  lado_b = x;
  
#suma_dos_lados = lado_a + lado_b

if lado_a + lado_b > lado_c:
  if lado_a == lado_b and lado_b == lado_c:
    print("es un triangulo equilatero")
  else:
    if lado_a != lado_b and lado_b != lado_c:
      print ("es un triangulo escaleno")
    else:
      print ("es un triangulo isoceles")
else:
  print("las medidas ingresadas no dibujan un triangulo")
```
